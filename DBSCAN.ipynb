```python
from sklearn.datasets import make_moons, make_circles
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Dataset 1: Moons
X_moons, _ = make_moons(n_samples=500, noise=0.05, random_state=42)
dbscan_moons = DBSCAN(eps=0.2, min_samples=5).fit_predict(X_moons)

# Dataset 2: Circles
X_circles, _ = make_circles(n_samples=500, factor=0.5, noise=0.05, random_state=42)
dbscan_circles = DBSCAN(eps=0.15, min_samples=5).fit_predict(X_circles)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.scatter(X_moons[:,0], X_moons[:,1], c=dbscan_moons, cmap='viridis')
ax1.set_title("DBSCAN: Moons")
ax2.scatter(X_circles[:,0], X_circles[:,1], c=dbscan_circles, cmap='viridis')
ax2.set_title("DBSCAN: Circles")
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler


iris = load_iris()
X_iris = StandardScaler().fit_transform(iris.data)

# Apply DBSCAN
dbscan_iris = DBSCAN(eps=0.6, min_samples=5)
labels = dbscan_iris.fit_predict(X_iris)

fig, ax = plt.subplots( figsize=(12, 4))
ax.scatter(X_iris[:,0], X_iris[:,1], c=labels, cmap='gnuplot')
ax.set_title("DBSCAN: iris")
plt.show()
```
