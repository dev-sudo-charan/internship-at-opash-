```python
import numpy as np
from scipy import stats

# Dataset representing ages of people in a room (with one outlier)
data = np.array([22, 25, 23, 22, 24, 25, 23, 22, 95])
```

```python
# Mean: Sum / Total count
mean_val = np.mean(data)

# Median: The middle value (automatically sorts internally)
median_val = np.median(data)

# Mode: The most frequent value
# stats.mode returns an object; we extract the first value
mode_result = stats.mode(data, keepdims=True)
mode_val = mode_result.mode[0]

print(f"Mean: {mean_val:.2f}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")
```

```python
# Variance: Average of squared differences from mean
variance_val = np.var(data)

# Standard Deviation: Square root of variance
std_dev_val = np.std(data)

print(f"Variance: {variance_val:.2f}")
print(f"Standard Deviation: {std_dev_val:.2f}")
```

```python
# Generate 1000 points following a Normal Distribution
# (Mean = 0, Std Dev = 1)
normal_data = np.random.normal(loc=0, scale=1, size=1000)

print(f"Normal Data Mean: {np.mean(normal_data):.4f}")
print(f"Normal Data Std Dev: {np.std(normal_data):.4f}")
```
