```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 1. Create 3 hidden groups of data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# 2. The Elbow Method: Finding the "True" K
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 4))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia') 
plt.show()
```

```python
import pandas as pd
import seaborn as sns

# Let's simulate mall customer data
data = {
    'Income': [15, 16, 17, 18, 70, 72, 80, 85, 20, 21, 75, 88],
    'Spending': [39, 81, 6, 77, 40, 45, 33, 90, 2, 95, 5, 82]
}
df = pd.DataFrame(data)

# 3. Apply K-Means (Assume K=4 from our domain knowledge)
kmeans = KMeans(n_clusters=4, random_state=42)
df['Segment'] = kmeans.fit_predict(df)

# Visualize the segments
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Income', y='Spending', hue='Segment', palette='viridis', s=300)
plt.title('Customer Segments: Finding the Big Spenders')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 1. Generate 100 random customer locations (x, y coordinates)
np.random.seed(42)
customers = np.random.rand(100, 2) 

# 2. Place 3 delivery drivers (Centroids)
kmeans = KMeans(n_clusters=4, n_init=10)
kmeans.fit(customers)

# 3. Get driver locations and customer assignments
drivers = kmeans.cluster_centers_
assignments = kmeans.labels_

# 4. Plot
plt.scatter(customers[:, 0], customers[:, 1], c=assignments, cmap='viridis', alpha=0.5)
plt.scatter(drivers[:, 0], drivers[:, 1], c='red', marker='X', s=200, label='Drivers')
plt.title("Pizza Driver Optimization")
plt.legend()
plt.show()
```

```python
# 1. Suppose a new customer orders at a specific location
new_customer = np.array([[0.1, 0.9]]) 

# 2. Use the 'predict' method to find the nearest driver
assigned_driver = kmeans.predict(new_customer)

print(f"New customer at {new_customer[0]} is assigned to Driver: {assigned_driver[0]}")

# 3. Visualizing where they fit
plt.scatter(customers[:, 0], customers[:, 1], c=assignments, cmap='viridis', alpha=0.3)
plt.scatter(drivers[:, 0], drivers[:, 1], c='red', marker='X', s=200)
plt.scatter(new_customer[:, 0], new_customer[:, 1], c='black', marker='*', s=300, label='New Order')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import make_moons

# 1. Create moon-shaped data
X, _ = make_moons(n_samples=200, noise=0.05, random_state=0)

# 2. Try K-Means (It will fail to separate the moons)
kmeans = KMeans(n_clusters=2)
labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title("K-Means Failing on Moon-Shaped Data")
plt.show()
```
