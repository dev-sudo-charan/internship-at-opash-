```python
import numpy as np

# create a 2x2 matrix
matrix1 = np.array([[1, 3], 
                   [5, 7]])

print("2x2 Matrix:\n",matrix1)

# create a 3x3  matrix
matrix2 = np.array([[2, 3, 5],
             	    [7, 14, 21],
                    [1, 3, 5]])
                    
print("\n3x3 Matrix:\n",matrix2)
```

```python
import numpy as np

# create two matrices
matrix1 = np.array([[1, 3]]
             		)
             
matrix2 = np.array([[2, 6, 8], 
                    [4, 8, 10]])

# calculate the dot product of the two matrices
result = np.dot(matrix1, matrix2)

print("matrix1 x matrix2: \n",result)

# note: you can multiply matrixes like [ 1*2 and 2*3 , as long as colomn of 1st matrix and row of second matrix are same you can multiply it.] , -by your's truely.
```

```python
import numpy as np

# create a matrix
matrix1 = np.array([[2, 6, 8], 
                    [4, 8, 10]])

# get transpose of matrix1
result = np.transpose(matrix1)

print(result)
```

```python
import numpy as np

# create a 3x3 square matrix
matrix1 = np.array([[1, 3, 5], 
             		[7, 9, 2],
                    [4, 6, 8]])

# find inverse of matrix1
result = np.linalg.inv(matrix1)

print(result)
```

```python
import numpy as np

# create a matrix
matrix1 = np.array([[1, 2, 3], 
             		[4, 5, 1],
                    [2, 3, 4]])

# find determinant of matrix1
result = np.linalg.det(matrix1)

print(result)
```

```python
import numpy as np

# create a 2x3 matrix
matrix1 = np.array([[1, 2, 3], 
             		[4, 5, 7]])

result = matrix1.flatten()
print("Flattened 2x3 matrix:", result)
```
