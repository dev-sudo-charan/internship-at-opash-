```python
import pandas as pd

# Load the real Titanic dataset from a public URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Let's see what we are dealing with
print("Dataset Shape:", df.shape)
print("\nMissing Values:")
print(df.isnull().sum())
print("\nFirst 5 rows:")
print(df.head())
```

```python
import matplotlib.pyplot as plt

# 1. Let's get the counts of Survivors vs Non-Survivors
survival_counts = df['Survived'].value_counts()

# 2. Create the plot
plt.figure(figsize=(8, 5)) # Set the size
plt.bar(['Died', 'Survived'], survival_counts, color=['red', 'green'])

# 3. Add the "Story" (Labels)
plt.title('Titanic Survival Counts')
plt.xlabel('Outcome')
plt.ylabel('Number of Passengers')

# 4. Display
plt.show()
```

```python
# 1. Get counts using Pandas
counts = df['Sex'].value_counts() # Results in: male: 577, female: 314

# 2. Plot
fig, ax = plt.subplots()
ax.bar(counts.index, counts.values, color=['blue', 'pink'])
ax.set_title("Gender Distribution on Titanic")
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Prepare Data
days = [1, 2, 3, 4, 5]
model_a_scores = [10, 25, 45, 70, 90] # Progresses fast
model_b_scores = [5, 15, 20, 35, 50]   # Progresses slow

# 2. Studio Setup
fig, ax = plt.subplots()

# 3. Draw the lines with LABELS (The Tags)
ax.plot(days, model_a_scores, label='High Learning Rate', color='blue', marker='o')
ax.plot(days, model_b_scores, label='Low Learning Rate', color='red', marker='s')

# 4. The Magic Step - Without this, the 'labels' above won't show up!
ax.legend(loc='best')
# 5. Add context
ax.set_title("AI Training Comparison")
ax.set_xlabel("Days")
ax.set_ylabel("Accuracy (%)")

plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Prepare the counts first (it makes the code cleaner)
gender_counts = df['Sex'].value_counts()

# 2. Setup the 1x2 studio
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# --- Left Graph: Bar Chart ---
# X = names, Y = counts
ax1.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'salmon'])
ax1.set_title("Gender Distribution")
ax1.set_ylabel("Count")

# --- Right Graph: Histogram ---
# Histograms only need 1 column (it counts the ranges for you!)
ax2.hist(df['Fare'], bins=20, color='gold', edgecolor='black')
ax2.set_title("Fare Distribution")
ax2.set_xlabel("Ticket Price")

plt.tight_layout()
plt.show()
```
