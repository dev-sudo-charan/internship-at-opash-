```python
import pandas as pd

# Load the real Titanic dataset from a public URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Let's see what we are dealing with
print("Dataset Shape:", df.shape)
print("\nMissing Values:")
print(df.isnull().sum())
print("\nFirst 5 rows:")
print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# You just point to the column name
sns.countplot(data=df, x='Sex')
plt.show()
```

```python
sns.countplot(data=df, x='Sex', hue='Survived')
sns.set_theme(style="ticks")
plt.show()
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title("Survival Rate by Ticket Class")
plt.show()
```

```python
# sns.histplot is the Seaborn version of a histogram
sns.histplot(data=df, x='Age', kde=True)
```

```python
# This creates a map where 'True' (missing) is one color and 'False' is another
sns.heatmap(df.isnull(), cbar=True, yticklabels=False, cmap='viridis')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Setting the theme for a professional look
sns.set_theme(style="whitegrid")

# 2. Creating the box plot
# We tell Seaborn to look at the 'Fare' column
sns.boxplot(data=df, x='Fare', color='lightcoral')

plt.title("Statistical Summary of Ticket Fares")
plt.show()
```

```python
# X is the Category (Class), Y is the Number (Fare)
sns.boxplot(data=df, x='Pclass', y='Fare', palette='Set2')
```

```python
# 1. Calculate the correlations (only for numeric columns)
# .corr() finds the relationships between all numbers
corr_matrix = df.corr(numeric_only=True)

# 2. Plot the Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")

plt.title("Correlation Heatmap of Titanic Features")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. We pick only the columns we care about for the AI
cols_to_plot = ['Survived', 'Pclass', 'Age', 'Fare']

# 2. The One-Liner
# hue='Survived' will color the dots: Blue for Died, Orange for Lived
sns.pairplot(df[cols_to_plot], hue='Survived', palette='husl')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# This gives you JUST the smooth line
sns.kdeplot(data=df, x='Age', fill=True, color="blue")
plt.show()
```

```python
# Comparing the age density of those who lived vs those who died
sns.kdeplot(data=df, x='Age', hue='Survived', fill=True, common_norm=False)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Using a Violin Plot: It's a KDE and a Box Plot combined!
sns.violinplot(data=df, x='Sex', y='Age', hue='Survived', split=True)

plt.title("Age & Sex Distribution of Survival")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# We set x as the categories and y as the number we want to measure
sns.violinplot(data=df, x='Pclass', y='Age', hue='Survived', split=True)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# The kind='reg' adds a regression line (the AI's 'best guess' line)
sns.jointplot(data=df, x='Age', y='Fare', kind='reg', color='teal')

plt.show()
```

```python
sns.jointplot(data=df, x='Age', y='Fare', kind='hex', color='purple')
```
